```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: np.exp(-t) #ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s [i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \ for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc="lower right")
plt.show()
```

```python
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
y0 = 1 # Initial Condition

# Define the differential equation y' = f(x, y)
def f(x, y):
    return x**2 - 4*y

# Explicit Euler Method
y = np.zeros(len(t))
y[0] = y0

for i in range(0, len(t) - 1):
    y[i + 1] = y[i] + h * f(t[i], y[i])

# Exact solution for comparison (if known)
def exact_solution(x):
    return (1/16)*(-3*np.exp(-4*x) + x**2 + (1/2)*x + (1/16))

# Print the value at x = 0.1
print("Approximate value at x = 0.1:", y[int(0.1 / h)])

plt.figure(figsize = (12, 8))
plt.plot(t, y, 'b--', label='Approximate')
plt.plot(t, exact_solution(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \ for Simple ODE')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend(loc="lower right")
plt.show()
```

```python
h = 0.1 # Step size
x = np.arange(0, 0.5 + h, h) # Numerical grid from 0 to 0.5
y0 = 1 # Initial condition

# Define the differential equation y' = f(x, y)
def f(y):
    return np.sin(y)

# Explicit Euler Method
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * f(y[i])
    y[i + 1] = round(y[i + 1], 4)  # Keeping four decimal places

# Print the approximate values for each step
print("Approximate values using Explicit Euler Method:")
for xi, yi in zip(x, y):
    print(f"x = {xi:.1f}, y = {yi:.4f}")

plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo--', label='Approximate')
plt.title('Approximate Solution for y\' = sin(y) using Explicit Euler Method')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.legend(loc="lower right")
plt.show()
```
